```python
def get_brand_name(car_name):
    car_name = car_name.split(' ')[0]
    return car_name.strip()
```

```python

# Calculate evaluation metrics
def calculate_metrics(y_true, y_pred):
    r2 = metrics.r2_score(y_true, y_pred)
    rmse = metrics.mean_squared_error(y_true, y_pred, squared=False)
    mae = metrics.mean_absolute_error(y_true, y_pred)
    mape = metrics.mean_absolute_percentage_error(y_true, y_pred)
    return r2, rmse, mae, mape

# Calculate metrics for each model
linear_r2, linear_rmse, linear_mae, linear_mape = calculate_metrics(y_test, predict)
lasso_r2, lasso_rmse, lasso_mae, lasso_mape = calculate_metrics(y_test, lasso_predict)
ridge_r2, ridge_rmse, ridge_mae, ridge_mape = calculate_metrics(y_test, ridge_predict)
rf_r2, rf_rmse, rf_mae, rf_mape = calculate_metrics(y_test, random_forest_predict)

# Create a table to display the results
metrics_table = pd.DataFrame({
    'Model': ['Linear Regression', 'Lasso Regression', 'Ridge Regression', 'Random Forest'],
    'R2': [linear_r2, lasso_r2, ridge_r2, rf_r2],
    'RMSE': [linear_rmse, lasso_rmse, ridge_rmse, rf_rmse],
    'MAE': [linear_mae, lasso_mae, ridge_mae, rf_mae],
    'MAPE': [linear_mape, lasso_mape, ridge_mape, rf_mape]
})

metrics_table
```
